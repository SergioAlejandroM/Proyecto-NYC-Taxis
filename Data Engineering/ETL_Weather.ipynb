```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Data/daily_weather_data.csv')
display(data)
```

```python
columns_to_drop = ['precipitation_sum','wind_gusts_10m_max']

df_clean = data.drop(columns_to_drop, axis=1)
```

```python
df_clean
```

```python
df_clean['rain_sum'] = np.where(df_clean['rain_sum'] >= 0.5, "Si", "No")

df_clean['snowfall_sum'] = np.where(df_clean['snowfall_sum'] >= 0.5, "Si", "No")
```

```python
df_clean.to_csv("../Data/Data Procesada/Weather.csv", index= False)
```
