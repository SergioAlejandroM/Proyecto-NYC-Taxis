```python
import pandas as pd
import glob
import os
```

```python
zonas_df = pd.read_csv('../Data/Data Procesada/Zonas.csv')
```

```python
for file in glob.glob('./transformed_files/*.csv'):
    df = pd.read_csv(file)

    merged_df = df.merge(zonas_df[['LocationID', 'Borough_ID']], left_on='PULocationID', right_on='LocationID', how='left')

    merged_df['PULocationID'] = merged_df['Borough_ID']

    merged_df.drop(columns=['Borough_ID', 'LocationID','pickup_time','dropoff_time'], inplace= True)

    merged_df.to_csv(file, index= False)
```

```python
# Directorio donde están los archivos CSV
file_path = './transformed_files/'  # Cambia esta ruta a tu directorio
output_path = '../Data/Data Procesada/Viajes/'  # Cambia esta ruta para guardar los archivos procesados

# Asegúrate de que el directorio de salida exista
os.makedirs(output_path, exist_ok=True)

# Leer todos los archivos CSV del directorio
all_files = glob.glob(file_path + "/*.csv")

# Procesar cada archivo individualmente
for filename in all_files:
    # Leer el archivo CSV
    df = pd.read_csv(filename)
    
    # Agrupar por PULocationID y Fecha, sumar trip_distance y fare_amount, y mantener Es_verde
    grouped_df = df.groupby(['PULocationID', 'Fecha'], as_index=False).agg({
        'trip_distance': 'sum',
        'fare_amount': 'sum',
        'Es_verde': 'first'
    })

    # Añadir la columna cantidad_viajes
    grouped_df['cantidad_viajes'] = df.groupby(['PULocationID', 'Fecha']).size().values

    # Crear el nombre del archivo de salida
    output_filename = output_path 
    
    # Guardar el DataFrame agrupado en un nuevo archivo CSV
    grouped_df.to_csv(filename, index=False)

    print(f'Archivo procesado guardado como: {filename}')
```
